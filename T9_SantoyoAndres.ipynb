Código de solución estudiante 1

```python
# Primero importamos todas las librerias, paquetes y/o funciones que vamos a utlizar
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# la funcion se comun para los tres casos, la definimos solo una vez
# Función f(y,t) que vamos a integrar, comun para todos los casos
def amortiguado(y,t,k,m,c):
    y1 = y[0]
    y2 = y[1]
    return np.array([y2, -(k/m)*y1-(c/m)*y2])
```

```python
# Especificamos los puntos de tiempo donde queremos la solución
# Es comun para los tres casos de movimientos, la definmos una vez
t = np.linspace(0, 30, 300)
```

```python
# Condiciones iniciales [1,1]

# *** Movimiento Amortiguado ***

# Definimos los parámetros k, m y c
k, m, c = 3, 1, 0.5

# Condiciones iniciales
y0 = np.array([1,1])
```

```python
# Simulacion
# Solución numérica
y = odeint(amortiguado, y0, t, args=(k,m,c))
# Extraer y1 y y2
y1 = y[:,0]
y2 = y[:,1]
```

```python
omega0=(k/m)**0.5
```

```python
omega0 = (k/m)**0.5
Gamma = c/(2*m)
```

```python
# *** Sobreamortiguado si se tiene  movimiento oscilatorio sobreamortiguado ***

# Nuevas constantes
k = .1 # Constante del muelle
m = 1.0   # Masa
c = 1 # Constante de amortiguación
```

```python
omega0 = np.sqrt(k/m)
Gamma = c/(2*m)
```

```python
# Simular
y = odeint(amortiguado, y0, t, args=(k,m,c))
y1s = y[:,0]
y2s = y[:,1]
```

```python
#***  Amortiguamiento crítico si se tiene movimiento críticamente amortiguado ***

# Nuevas constantes
k = .0625 # Constante del muelle
m = 1.0   # Masa
c = .5 # Constante de amortiguación
```

```python
omega0 = np.sqrt(k/m)
Gamma = c/(2*m)
```

```python
# Simular
y = odeint(amortiguado, y0, t, args=(k,m,c))
y1c = y[:,0]
y2c = y[:,1]
```

```python
tt = t
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, sharex='col',
                                                       sharey='row',figsize =(10,6))
ax1.plot(tt, y1, c = 'k')
ax1.set_title('Amortiguado', fontsize = 14)
ax1.set_ylabel('Posición', fontsize = 14)
ax2.plot(tt, y1s, c = 'b')
ax2.set_title('Sobreamortiguado', fontsize = 14)
ax3.plot(tt, y1c, c = 'r')
ax3.set_title('Crítico', fontsize = 16)
ax4.plot(tt, y2, c = 'k')
ax4.set_ylabel('Velocidad', fontsize = 14)
ax4.set_xlabel('tiempo', fontsize = 14)
ax5.plot(tt, y2s, c = 'b')
ax5.set_xlabel('tiempo', fontsize = 14)
ax6.plot(tt, y2c, c = 'r')
ax6.set_xlabel('tiempo', fontsize = 14)
plt.show()
```
